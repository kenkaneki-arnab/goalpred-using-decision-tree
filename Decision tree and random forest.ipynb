```python
from sklearn.model_selection import train_test_split
train_df,temp_df=train_test_split(df,test_size=0.3,random_state=42)
val_df,test_df=train_test_split(temp_df,test_size=0.5,random_state=42)
```

```python
train_target = train_df[["goals"]]
val_target=val_df[["goals"]]
test_target=test_df[["goals"]]
```

```python
train_input = train_df[["shots", "xGoals", "xGoalsChain","xGoalsBuildup"]]
val_input = val_df[["shots", "xGoals", "xGoalsChain","xGoalsBuildup"]]
test_input = test_df[["shots", "xGoals", "xGoalsChain","xGoalsBuildup"]]
```

```python
train_input
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(random_state=42)
```

```python
model.fit(train_input,train_target)
```

```python
#Regularisation
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
def max_depth_error(md):
    model=DecisionTreeClassifier(max_depth=md,random_state=42)
    model.fit(train_input,train_target)
    train_acc=1-model.score(train_input,train_target)
    val_acc=1-model.score(val_input,val_target)
    return{'Max Depth':md,'Train Error':train_acc,'Val Error':val_acc}
```

```python
for md in range (1,20):
    print(max_depth_error(md))
```

```python
#at max_depth=5 val_error is minimum hence the optimal model for us would be to take max depth as 5
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    "max_depth": [3, 5, 7, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "criterion": ["gini", "entropy"]
}

# Grid Search
grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,             # 5-fold cross validation
    scoring='accuracy',
    n_jobs=-1         # uses all CPU cores
)

# Fit on your df
grid.fit(train_input, train_target)

# Best model found
best_model = grid.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def evaluate_best_model(best_model, test_input, test_target):


    # Predict
    y_pred = best_model.predict(test_input)

    # Accuracy
    acc = best_model.score(test_input, test_target)
    print("🔥 Test Accuracy:", acc)

    # Confusion Matrix
    cm = confusion_matrix(test_target, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title("Confusion Matrix - Best Model")
    plt.show()

    return y_pred
```
